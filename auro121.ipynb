```python
import xml.etree.ElementTree as ET
tree = ET.parse('../input/orders/orders.xml')
root = tree.getroot()
```

```python
def myFunc(e):
    return e.price
B=[]
class order:
    def __init__(self,type,price,vol,id):
        self.type = type
        self.price = price
        self.volume = vol
        self.id = id
    def __str__(self): 
        return "%d@%.2f"%(self.volume, self.price) 
class Book:
    def __init__(self, name):
        self.name = name
        self.buy=[]
        self.sell=[]
        #B.append(self)
    def add_order(self, order):
        if(order.type=='SELL'):
            for i in range(len(self.buy)):
                if(order.price>self.buy[i].price):
                    self.buy[i-1]
                    if(self.buy[i-1].volume>=order.volume):
                        self.buy[i-1].volume-=order.volume
                    else:
                        t = order
                        t.volume-=self.buy[i-1].volume
                        self.sell.append(t)
                        del self.buy[i-1]
            self.sell.append(order)
        else:
            self.buy.append(order)
        self.buy.sort(key=myFunc)
        self.sell.sort(key=myFunc)
        
def del_order(bookid,orderid):
    for i in range(len(B)):
        if(B[i].name==bookid):
            for j in range(len(B[i].sell)):
                if(B[i].sell[j].id==orderid):
                    del B[i].sell[j]
                    break
            for j in range(len(B[i].buy)):
                if(B[i].buy[j].id==orderid):
                    del B[i].buy[j]
                    break
```

```python
from decimal import Decimal
precision=2
i=0
for child in root:
    i+=1
    print(child.tag)
    print(child.attrib)
    for att,val in child.attrib.items():
        if(child.tag=='AddOrder'):
            flag=False
            if(att=='book'):
                for x in B:
                    if(x.name==val):
                        flag=True
                        x.add_order(order(child.attrib.get('operation'),float(child.attrib.get('price')),int(child.attrib.get('volume')),child.attrib.get('orderId')))
                        break;
                if(not flag):
                    Bk = Book(val)
                    Bk.add_order(order(child.attrib.get('operation'),float(child.attrib.get('price')),int(child.attrib.get('volume')),child.attrib.get('orderId')))        
                    B.append(Bk)
                    break
        if(child.tag=='DeleteOrder'):
            del_order(child.attrib.get('book'),child.attrib.get('orderId'))
            
    if(i==10):
        break;
print("done")
```

```python
import time
def print_out():
    current_time = time.ctime()
    print(current_time)
    for j in range(len(B)):
        print(B[j].name)
        print("buy\t|\t sell")
        #for i in range(len(B[j].buy)):
        #    print(B[j].buy[len(B[j].buy)-i-1])
        #print("sell")
        #for i in range(len(B[j].sell)):
        #    print(B[j].sell[i])
        k=0
        for b,s in zip(reversed(B[j].buy),B[j].sell):
            print(b,"\t|\t",s)
            k+=1
        if(len(B[j].buy)>len(B[j].sell)):
            for i in range(k,len(B[j].buy)):
                print(B[j].buy[len(B[j].buy)-i-1],"\t|\t")
        else:
            for i in range(k,len(B[j].sell)):
                print("\t|\t",B[j].sell[i])
        print("\n")
    current_time = time.ctime()
    print(current_time)
```
